# Handling Missing values in Pandas


*   Pandas intepret missing values as NaN (not a value)




```python
#you can pass what type of value you want to consider as NaN in na_value=
df = pd.read_csv("/content/1. Regression - Module - (Housing Prices).csv", na_values='Fair') 
df.head()
```

```python
#you can pass list of values 
df = pd.read_csv("/content/1. Regression - Module - (Housing Prices).csv", na_values=['Fair','Excellent'] )
df.head()
```

```python
df = pd.read_csv("/content/1. Regression - Module - (Housing Prices).csv" )
df.isnull() # using isnull(), if there is any null value it will represent by True
```

```python
df.isnull().sum() #this will give how many null values are present in each column
```

```python
df.isnull().sum().sum()  #this will give total null values in a DF
```

```python
df.notnull() # using notnull(), not null values will represent by True
```

```python
import numpy as np
sr = pd.Series(['a','b','c',np.nan,'d',np.NaN])
sr
```

```python
sr.isnull().sum()
```

```python
df.dropna(axis=1) #drops/removes rows which conntain NaN values in any of the column
# df.head()
```

```python
df.dropna(how='any') #it will remove row/column if there is any nan values
```

```python
df.dropna(how='all') #if all values in row or columns, than only it will remove that row/column
```

```python
df.dropna(axis=1,thresh=0)
```

```python
df.dropna(subset=['ID']) #this will remove the particular column which is containg nan value in ID column
df.dropna(subset=['No of Bedrooms']) #this will remove the particular column which is containg nan value in No of Bedrooms column
```

```python
df.dropna(inplace=True) #this will update original dataframe, and will not create new dataframe
df
```

```python
# fillna()method fills the NaN values with given(input) values
df = pd.read_csv("/content/1. Regression - Module - (Housing Prices).csv" )
df.fillna(0)# this is will fill all the NaN values with 0
```

```python
df.fillna({'ID':0, 'No of Bathrooms':2})
```

```python
df.fillna(method='ffill') #this will fill previous value of column in NaN values
```

```python
df.fillna(method='bfill') #this will fill value of next row in NaN values
```

```python
df.fillna(0, limit=1) # this will fill only one Nan value in a column
df.fillna(0, limit=5) # this will fill only five Nan value in a column
```

```python
# replace() : values to DataFrame are replaced with other values dynamically
df.replace(to_replace='No', value="Hello") # to_replace= value you want to replace
                                            # value = replacing value
```

```python
df.replace({'No of Bedrooms':[1,2,3]}, 'none')
```

```python
df.replace('[A-Za-z]',0,regex=True)
```
