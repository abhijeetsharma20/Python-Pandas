pandas provide various facilities for easily combining together Series, Dataframe and panel objects

merge ( ) : pandas merge function connects or merge column or indexes in DataFrame based on one or more keys.

```python
df1 = pd.DataFrame({'Id' : [1,2,3,4], 'Name' : ['Abhijeet','Rishita','Shailendra',"arwaz"]})
df2 = pd.DataFrame({'Id' : [1,2,3,5], 'Age' : [1,2,3,4]})
```

```python
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,on="Id")
```

```python
pd.merge(df1,df2,on="Id", how='inner')
```

```python
pd.merge(df1,df2,on="Id", how='left')
```

```python
pd.merge(df1,df2,on="Id", how='right')
```

```python
pd.merge(df1,df2,on="Id", how='outer')
```

```python
pd.merge(df1,df2,on="Id", how='outer', indicator=True)
```

```python
df1 = pd.DataFrame({'Id' : [1,2,3,4], 'Name' : ['Abhijeet','Rishita','Shailendra',"arwaz"]})
df2 = pd.DataFrame({'Id' : [5,6,7,8], 'Age' : [1,2,3,4]})
```

```python
pd.merge(df1,df2,left_index=True, right_index=True)
```

```python
pd.merge(df1,df2)
```

```python
df1 = pd.DataFrame({'Id' : [1,2,3,4], 'class' : ['Abhijeet','Rishita','Shailendra',"arwaz"]})
df2 = pd.DataFrame({'Id' : [1,2,3,5], 'class' : [1,2,3,4]})
pd.merge(df1,df2,on='Id',suffixes=("_left",'_right'))
```

Concat( ) 

```python
df1 = pd.Series([1,2,3])
df2 = pd.Series([4,5,6])
pd.concat([df1,df2])
```

```python
df1 = pd.DataFrame({'Id':[1,2,3,4],'class':[5,6,7,8]})
df2 = pd.DataFrame({'Id':[5,6], 'class':[11,12]})
pd.concat([df1,df2],axis=1,keys=['df1','df2'])
```

```python
df1 = pd.DataFrame({'Id':[1,2,3,4],'class':[5,6,7,8]})
df2 = pd.DataFrame({ 'Marks':[11,12]})
pd.concat([df1,df2],axis=1)
```

join( ): DataFrame join is a covinient for combining the column of two potentially different indexed.

```python
df1 = pd.DataFrame({'Id':[1,2,3,4],'class':[5,6,7,8]}, index=['a','b','c','d'])
df2 = pd.DataFrame({'Id':[1,2],'class':[5,6]},index=['a','b'])
display(df1,df2)
```

```python
df1.join(df2, lsuffix='_1', rsuffix='_2')
```

```python
df1 = pd.DataFrame({'Id':[1,2,3,4],'class':[5,6,7,8]}, index=['a','b','c','d'])
df2 = pd.DataFrame({'I':[1,2],'class':[1,2]},index=['a','b'])
df1.append(df2,ignore_index=True, sort=True)
```

pivot table : the levels in the pivot table will be stored in a multiindex objects (hierachial index) on the index and columns of result dataFrame

```python
df = pd.read_csv('/content/1. Regression - Module - (Housing Prices).csv')
df.pivot_table(index='No of Bedrooms', columns='Sale Price', aggfunc='max', fill_value=0, margins=True)
```
